```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-colorblind')
```

```python
np.random.seed(123)

df = pd.DataFrame({'A': np.random.randn(365).cumsum(0),
                  'B': np.random.randn(365).cumsum(0) + 20,
                  'C': np.random.randn(365).cumsum(0) - 20},
                 index=pd.date_range('2017-01-01', periods=365))

df.head()
```

```python
df.plot();
```

```python
df.plot('A', 'C', kind='scatter');
```

```python
df.plot.scatter('A', 'C');
```

```python
df.plot.scatter('A', 'C', c='B', colormap='viridis');
```

```python
df.plot.box();
```

```python
df.plot.hist(alpha=0.7);
```

```python
df.plot.kde(); # kernel density estimation
```
