## 1. Online Courses

To build basic understanding of ML, online courses are the best place to start.

[Machine Learning](https://www.coursera.org/learn/machine-learning) (Andrew Ng)


The best part of Andrew's calss is that he skips complicated math equations but gives you the most important **intuition**.  

Summary of what you learnt:
- **Regression** is for predicting continuous data.
- **Classification** is for giving unknown data a pre-defined label.
- In **Supervised Learning** you have data with corresponding pre-defined label for training.
- In **Unsupervised Learning** you only have data, and need ML to help you do clustering.
- **Cost Function** is what you defined to tell ML "how bad is currect predicted label comparing with real label"
- **Gradient Descent** is the machanism to optimize your cost function, based on Deriative. 
- **Sigmoid** function is the math equation used to map any value to [0-1]
- **Overfitting** is that the function learnt is too good for training data, but performs poorly for unknown data.
- **Regularization** is a mathematical way to restrict complex parameters, to avoid overfitting.
- **SVM** is based on solid math to find the best margin separating.
- **Neural Network** forward and backward propagation, intuitively.

## 3. Try it in Python

Python is probably the easiest programming language providing comprehensive ML libraries. The ML libraries in Python is roughly like this (not very precise)



Let's focus on the left side of the figure. What you need? Yes **scikit learn**!

The following code example trains a SVM model to classify Iris dataset.

```python
# --------------------------------------------------
# Load pre-saved Iris dataset
# --------------------------------------------------
from sklearn import datasets


iris = datasets.load_iris()
X = iris['data']       # the features of each sample
y = iris['target']   # the label of each sample

print('total data count = {}'.format(len(X)))
print('total label count = {}'.format(len(y)))
print('distinct label count = {}'.format(len(set(y))))

print('first data = {}'.format(X[0]))  # Sepal Length, Sepal Width, Petal Length and Petal Width
print('first label = {}'.format(y[0]))

# --------------------------------------------------
# Split the data to Train / Test
# --------------------------------------------------
from sklearn.model_selection import train_test_split

# separate our data into randomized train/test parts
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=99)

print('train data size = {}'.format(len(X_train)))
print('test data size = {}'.format(len(X_test)))

# --------------------------------------------------
# Now we use SVM to do classification based on data
# --------------------------------------------------
from sklearn.svm import SVC

clf = SVC(kernel='linear')
clf.fit(X_train, y_train)  # this is all you need for training

print(clf)

print('train score = {}'.format(clf.score(X_train, y_train)))
print('test score = {}'.format(clf.score(X_test, y_test)))
```

## 4. Principal Component Analysis (PCA)

The features in Iris dataset are pre-defined and very simple. For real data, you might need to do feature engineering to get the data part to training. Say, you want to train a model able to recognize authorship. You may do feature engineering of an article and get features such as: 

- ratio_word_per_sentence
- ratio_comma_per_paragraph
- word count per sentence
- word count per paragraph
- newline count per paragraph
- ...

How do you know if the features you choose do really able to classify your data into different labels?

One way to do this is to plot them on graph, and check if the samples on graph can be separated visually. Because it's not easy to plot graph more than 3 dimemsions, we need a method to reduce the dimensionality. You may hear about Principal Component Analysis (PCA) and know it can help. Let's see PCA using Iris example.


```python
# this is a magic command for jupyter notebook to plot picture inline instead of opening a new window
%matplotlib inline
```

```python
import pylab as pl
from itertools import cycle

def plot_PCA_2D(data, target, target_names):
    if len(target_names) == 2:
        colors = cycle('rg')
    elif len(target_names) == 3:
        colors = cycle('rgb')
    else:
        return
    
    target_ids = range(len(target_names))

    pl.figure()
    
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], 
                   data[target == i, 1],
                   c=c, label=label)
    pl.legend()
```

```python
# --------------------------------------------------
# Load Iris dataset
# --------------------------------------------------
from sklearn import datasets

iris = datasets.load_iris()
X = iris['data']       # the features of each sample
y = iris['target']   # the label of each sample

# --------------------------------------------------
# Plot it
# --------------------------------------------------
import matplotlib as plt
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print(X_pca[:5])

plot_PCA_2D(X_pca, y, ['0', '1', '2'])
```
